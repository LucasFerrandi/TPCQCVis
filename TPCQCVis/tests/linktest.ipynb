```python
from IPython.display import HTML

link_url = "/reports/2023/LHC23zc/TPC_AQC_LHC23zc_cpass0.html #Tracks"
link_text = "HOVER"

html_code = f'''
<style>
#ifrm1 {{
  display: none;
}}
</style>

<a href="{link_url}" onmouseover="showIframe('{link_url}')" onmouseout="hideIframe()">{link_text}</a>
<iframe id='ifrm1' name='ifrm1' src='' style='width:400px; height:400px;' onmouseover="showIframe('{link_url}')" onmouseout="hideIframe()"></iframe>

<script>
function showIframe(url) {{
  document.getElementById('ifrm1').src = url;
  document.getElementById('ifrm1').style.display = 'block';
}}

function hideIframe() {{
  document.getElementById('ifrm1').style.display = 'none';
}}
</script>
'''

HTML(html_code)
```

```python
from IPython.display import HTML

link_url = "https://days.to/until/christmas"
link_text = "HOVER"

html_code = f'''
<style>
#ifrm1 {{
  position: absolute;
  display: none;
  width: 100%;
  height: 400px;
  top: 0;
  left: 0;
  z-index: 9999;
}}
.link-container {{
  position: relative;
  display: inline-block;
}}
</style>

<div class="link-container">
  <a href="{link_url}" onmouseover="showIframe('{link_url}')" onmouseout="hideIframe()">{link_text}</a>
</div>

<script>
function showIframe(url) {{
  var linkContainer = document.getElementsByClassName('link-container')[0];
  var linkPosition = linkContainer.getBoundingClientRect();
  
  var iframe = document.getElementById('ifrm1');
  iframe.src = url;
  iframe.style.display = 'block';
  iframe.style.top = linkPosition.bottom + 'px';
  iframe.style.left = linkPosition.left + 'px';
}}

function hideIframe() {{
  document.getElementById('ifrm1').style.display = 'none';
}}
</script>

<iframe id='ifrm1' name='ifrm1' src=''></iframe>
'''

HTML(html_code)
```
